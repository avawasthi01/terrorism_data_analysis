# Problem Statement :
Find value of killed column only where country == ‘United States’?
Print 0 in place of missing values.
Print count of Killed as integer value.

```python
import csv
import numpy as np
with open('terrorismData.csv') as file_obj:
    rec=csv.DictReader(file_obj,skipinitialspace=True)
    rec=list(rec)
    country=[]
    killed=[]
    for i in rec:
        country.append(i['Country'])
        killed.append(i['Killed'])
    for i in range(len(killed)):
        if killed[i] == '':
            killed[i] = '0'
    country=np.array(country)
    
    killed=np.array(killed,dtype=float)
    killed=np.array(killed,dtype=int)
    for i in killed[np.where(country=='United States')]:
        print(i)
 
```

# Problem Statement :
Find total number of people killed from USA?
Note: Some columns of killed are empty so replace them by 0.
Print count of Killed as integer value

```python
import numpy as np
import csv
with open('terrorismData.csv') as file_obj:
    records=csv.DictReader(file_obj)
    records=list(records)
    killed=[]
    for i in records:
        if i['Country'] == 'United States':
            killed.append(i['Killed'])
    killed=np.array(killed)
    killed[killed == ''] ='0'
    killed=np.array(killed,dtype=float)
    print(int(np.sum(killed)))
            
```

# Problem Statement :
Find the number of attack held between day 10 and day 20?(ignoring the year and month)(including both day)
Print count of NumberOFAttack as integer value.

```python
import numpy as np
import csv
with open('terrorismData.csv') as file_obj:
    rec=csv.DictReader(file_obj)
    attack=[]
    rec=list(rec)
    for i in rec:
        if int(i['Day'])>=10 and int(i['Day'])<=20:
            attack.append(i['AttackType'])
    print(len(attack))
```

# Problem Statement :
Find the number of attack held between 1 Jan 2010 and 31 Jan 2010?(including both date).
Note Ignore the case where day is 0
Print count of NumberOFAttack as integer value.

```python
import csv
import numpy as np
with open("terrorismData.csv") as file_obj:
    file_data=csv.DictReader(file_obj,skipinitialspace=True)
    rec=list(file_data)
    attacks=[]
    for i in rec:
        if int(i['Year']=='2010') and i['Month'] == '1':
            attacks.append(i['Day'])
    x=np.array(attacks,dtype=float)
    print(len(x[(x<=31) & (x>=1)]))
```

# Problem Statement :
Find the casualty in the Red Corridor States ? Mainly Red corridor states include Jharkhand, Odisha, Andhra Pradesh, and Chhattisgarh.
Note: Casualty=Killed +Wounded
Print count of Casualty as integer value.

```python
import csv
import numpy as np
with open('terrorismData.csv') as file_obj:
    file_data=csv.DictReader(file_obj,skipinitialspace=True)
    casualty=[]
    for row in file_data:
        if(row['State']=='Jharkhand' or row['State']=='Odisha' or row['State']=='Andhra Pradesh' or row['State']=='Chhattisgarh'):
            casualty.append([row['Killed'],row['Wounded']])
    np_cas=np.array(casualty)
    np_cas[np_cas=='']='0.0'
    np_cas=np.array(np_cas,dtype=float)
    np_cas=np.array(np_cas,dtype=int)
    p=np.sum(np_cas,axis=1)
    np_max=np.sum(np_cas)
    print(np_max)
```

# Problem Statement :
Find the most frequent day of attack in a terrorismDataset ?
Note: Here np.unique can be used.
Print count of frequent day and number of attack as Integer value.

```python
import csv
import numpy as np
with open("terrorismData.csv") as obj:
    file_data=csv.DictReader(obj,skipinitialspace=True)
    day=[]
    for row in file_data:
        day.append(row['Day'])
    np_day=np.array(day)
    x,y=np.unique(np_day,return_counts=True)
    print(x[np.argmax(y)],np.max(y))
```
